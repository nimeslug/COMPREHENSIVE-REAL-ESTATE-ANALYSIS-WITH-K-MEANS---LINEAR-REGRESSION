# Veri Temizleme

```python
## NAME DEĞERİ ÇEKİLECEK SÜTUNLAR

import ast

# json formatında Name bilgisi olan sütunlardan Name i çekmek için fonksiyon
def extract_name(x):
    try:
        # Eğer x zaten bir dict değilse önce dict'e çevir, yoksa direkt döndür
        if isinstance(x, str) and x.strip().startswith('{'):
            return ast.literal_eval(x)['name']
        else:
            return x
    except:
        return x

#Name bilgisini çekmek istediğimiz sütunlara uyguladım:
df['district'] = df['district'].apply(extract_name)
df['county'] = df['county'].apply(extract_name)
df['residence'] = df['residence'].apply(extract_name)
df['heating'] = df['heating'].apply(extract_name)
df['fuel'] = df['fuel'].apply(extract_name)
df['build'] = df['build'].apply(extract_name)
df['buildState'] = df['buildState'].apply(extract_name)


df.head()
```

```python
## [] İÇİNDEN VERİ ÇIKARMA (ODA SAYILARI)

def remove_brackets(x):
    try:
        return float(ast.literal_eval(x)[0])
    except:
        return None

df['room'] = df['room'].apply(remove_brackets)
df['livingRoom'] = df['livingRoom'].apply(remove_brackets)
#bathroom floata çevirme
df['bathRoom'] = pd.to_numeric(df['bathRoom'], errors='coerce').astype(float)

df.head()
```

```python
## METREKARE ÇIKARIMI - 2 YENİ SÜTUN
import json

# sqm sütunundan netSqm ve grossSqm değerlerini çıkarmak için fonksiyon
def extract_sqm(x):
    try:
        d = json.loads(x.replace("'", '"'))  # Tek tırnakları düzelt
        net = d.get('netSqm', None)
        gross_list = d.get('grossSqm', [])
        gross = gross_list[0] if isinstance(gross_list, list) and gross_list else None

        # Float'a çevirdim
        net = float(net) if net is not None else None
        gross = float(gross) if gross is not None else None

        return net, gross
    except Exception as e:
        print(f"Hata: {e} --> {x}")
        return None, None

# Fonksiyonu kullanıp yeni net ve brüt sütunları tanımladım
df[['netSqm', 'grossSqm']] = df['sqm'].apply(lambda x: pd.Series(extract_sqm(x)))
df.drop('sqm', axis=1, inplace=True) # eski sütunu sildim
df.head()
```

```python
## SUBKATEGORİ ÇIKARIMI
def extract_type_name(x):
    try:
        # Eğer x string ve { ile başlıyorsa dict'e çevir
        if isinstance(x, str) and x.strip().startswith('{'):
            return ast.literal_eval(x)['typeName']
        # Eğer x zaten dict ise doğrudan typeName'ı al
        elif isinstance(x, dict):
            return x.get('typeName')
        else:
            return None
    except:
        return None

# Fonksiyonu uygula
df['subCategory'] = df['subCategory'].apply(extract_type_name)

df.head()
```

```python
# FLOOR AYRILDI

# JSON string'leri sözlüğe çevir
df_parsed = df["floor"].apply(lambda x: json.loads(x) if pd.notna(x) else {})

# Sözlükleri DataFrame'e aktar
floor_df = pd.json_normalize(df_parsed)

# Yeni sütunlar
df["bina_kat_sayisi"] = pd.to_numeric(floor_df["count"], errors="coerce").astype(float)
df["daire_kati"] = pd.to_numeric(floor_df["name"].str.extract(r"(\d+)")[0], errors="coerce").astype(float)

# Eski floor sütununu sil
df.drop('floor', axis=1, inplace=True)

# İsteğe bağlı: sonucu görmek için
df.head()
```

```python
# Sides sütununda yazan 4 farklı cephe seçeneği için 4 yeni sütun oluşturdum
# One hot encoding yapmış oldum 1 ve 0 değerleri atadım
import random
random.seed(54)

# Yönler listesi ve küçük harfli sütun isimleri
yonler = ['Kuzey', 'Güney', 'Doğu', 'Batı']
sütunlar = ['kuzey', 'guney', 'dogu', 'bati']

# Başlangıçta tüm yeni sütunları 0 ile oluştur
for s in sütunlar:
    df[s] = 0

def parse_and_assign(row):
    try:
        sides_list = ast.literal_eval(row)
    except:
        sides_list = []
    mevcut_cepheler = [item['name'].lower() for item in sides_list if 'name' in item]

    if mevcut_cepheler:
        # Cephe varsa 1 yaz
        return [1 if yon.lower() in mevcut_cepheler else 0 for yon in yonler]
    else:
        # Boşsa rastgele en az 1 cephe seç
        secilecek_sayisi = random.randint(1, len(yonler))
        rastgele_cepheler = random.sample(sütunlar, secilecek_sayisi)
        return [1 if s in rastgele_cepheler else 0 for s in sütunlar]

# Uygula ve sütunlara ata
df[sütunlar] = df['sides'].apply(parse_and_assign).tolist()
df.drop(columns=['sides'], inplace=True)
df.head()
```

```python
## Boş listelerin dönüşümü, boş verilerin NaN a çevrilmesi
import numpy as np

def convert_empty_dict_to_nan(x):
    try:
        if isinstance(x, str) and x.strip().startswith('{'):
            # null -> None yap
            x = x.replace('null', 'None')
            d = ast.literal_eval(x)
            if d.get('id') == 0 and d.get('name') is None:
                return np.nan
        return x
    except:
        return x

df['residence'] = df['residence'].apply(convert_empty_dict_to_nan)
df['fuel'] = df['fuel'].apply(convert_empty_dict_to_nan)
df['build'] = df['build'].apply(convert_empty_dict_to_nan)
df['buildState'] = df['buildState'].apply(convert_empty_dict_to_nan)

# Boş hücrelerin NaN yapılması

df = df.replace('', np.nan)
df.head()
```

```python
## Residence NaNlarını metrekare bilgisine göre doldurma:
## Mantığım: Metrekaresi büyükse muhtemelen dublekstir, düşükse dairedir gibi.
## Bunun için residence'ı metrekare ortalamasına göre gruplandırdım.

# Residence türlerine göre netSqm ortalamalarını hesapladım
residence_netSqm_mean = df.groupby('residence')['netSqm'].mean()

# Residence sütununda NaN olan satırlarda residence tahmini yapmak için fonksiyon
def fill_residence(row):
    if pd.isna(row['residence']):
        # Eğer residence eksikse, netSqm değerine en yakın ortalama değeri bul
        if pd.notna(row['netSqm']):
            # Ortalamalarla farkı hesapla
            closest_type = (residence_netSqm_mean - row['netSqm']).abs().idxmin()
            return closest_type
        else:
            return np.nan  # netSqm de yoksa boş bırak
    else:
        return row['residence']

df['residence'] = df.apply(fill_residence, axis=1)

df['residence'].value_counts(dropna=False) # Hangi residence tipinden kaç tane oldu
```

```python
#bina kat sayisi sütununun max ve 2.max değerini bulduk.(314 gürültülü veri kaldırılmalı!)
unique_sorted = sorted(df['bina_kat_sayisi'].unique())

# En büyük ve bir önceki en büyük değeri al
max_val = unique_sorted[-1]
prev_max = unique_sorted[-2]

print("Maksimum bina_kat_sayisi:", max_val)
print("Bir önceki maksimum (2. büyük):", prev_max)
```

```python
#314 max değeri gürültülü veri olduğundan ikinci max değeri(10) ile değiştirdik.
df.loc[df['bina_kat_sayisi'] == 314, 'bina_kat_sayisi'] = 10
```

```python
#daire kati bina katsayisindan büyük olamaz eşit olabilir küçük olabilir aykırı değer kontrolü yaptık.(Aşağıda tripleks villa için daire katini 1 ayarlayınca bu aykırı değer düzeldi.)
# daire_kati > bina_kat_sayisi olan satırları filtrele
violations = df[df['daire_kati'] > df['bina_kat_sayisi']]

# Eğer varsa yazdır
if not violations.empty:
    print("Kuralı ihlal eden satırlar (daire_kati > bina_kat_sayisi):")
    print(violations[['bina_kat_sayisi', 'daire_kati']])
else:
    print("Kuralı ihlal eden satır yok.")
```

```python
#hangi residence değerleri hangi subcategory değerlerine denk geliyor baktık ki ona göre daire_kati sütununu dolduralım.
# İlgili residence türlerini listele
residence_list = [
    'Daire', 'Tripleks', 'Çatı Dubleksi', 'Müstakil Ev',
    'Dubleks', 'Bahçe Dubleksi', 'İkiz Ev', 'Ters Dubleks'
]

# residence değeri bu listede olanları filtrele
filtered_df = df[df['residence'].isin(residence_list)]

# subCategory'ye göre gruplama ve sayma
grouped = filtered_df.groupby(['residence', 'subCategory']).size().reset_index(name='adet')

# Sonuçları yazdır
print(grouped)
```

```python
# İmarlı - Konut ve Konut+Ticaret için bina_kat_sayisi ve daire_kati değerlerini 0 yap
df.loc[df['subCategory'].isin(['İmarlı - Konut', 'Konut+Ticaret']), ['bina_kat_sayisi', 'daire_kati']] = 0
```

```python
# residence = Daire ve subCategory değerleri belirtilen değerlerse daire_kati'yi 1 yap
df.loc[
    (df['residence'] == 'Daire') &
    (df['subCategory'].isin(['Çiftlik Evi', 'Yazlık', 'Villa', 'Müstakil Ev'])),
    'daire_kati'
] = 1
```

```python
# residence değeri Bahçe Dubleksi olanların daire_kati değerini 1 yap
df.loc[df['residence'] == 'Bahçe Dubleksi', 'daire_kati'] = 1
```

```python
# Dubleks & (Bungalov, Müstakil Ev, Villa, Yazlık) için daire_kati = 1
df.loc[
    (df['residence'] == 'Dubleks') &
    (df['subCategory'].isin(['Bungalov', 'Müstakil Ev', 'Villa', 'Yazlık'])),
    'daire_kati'
] = 1
```

```python
# residence değeri Müstakil Ev olanların daire_kati değerini 1 yap
df.loc[df['residence'] == 'Müstakil Ev', 'daire_kati'] = 1
```

```python
# Ters Dubleks & (Müstakil Ev, Villa) için daire_kati = 1
df.loc[
    (df['residence'] == 'Ters Dubleks') &
    (df['subCategory'].isin(['Müstakil Ev', 'Villa'])),
    'daire_kati'
] = 1
```

```python
# Tripleks & (Çiftlik Evi, Müstakil Ev, Villa, Yazlık, Köşk) için daire_kati = 1
df.loc[
    (df['residence'] == 'Tripleks') &
    (df['subCategory'].isin(['Çiftlik Evi', 'Müstakil Ev', 'Villa', 'Yazlık', 'Köşk'])),
    'daire_kati'
] = 1
```

```python
# Çatı Dubleksi & (Prefabrik, Müstakil Ev, Villa, Yazlık) için daire_kati = 1
df.loc[
    (df['residence'] == 'Çatı Dubleksi') &
    (df['subCategory'].isin(['Prefabrik', 'Müstakil Ev', 'Villa', 'Yazlık'])),
    'daire_kati'
] = 1
```

```python
# İkiz Ev & (Müstakil Ev, Villa) için daire_kati = 1
df.loc[
    (df['residence'] == 'İkiz Ev') &
    (df['subCategory'].isin(['Müstakil Ev', 'Villa'])),
    'daire_kati'
] = 1
```
